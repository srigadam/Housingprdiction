```python
# i will use iloc[ row index start : end, coloumn index start : end ]
# features
x = df.iloc[ : , 0 : 13]
x.head()
```

```python
# outcome ....
y= df.iloc[:, 13:]
```

```python
from sklearn.model_selection import train_test_split
# scikit learn (sklearn is the library whiich resposible for model creation)
# syntax : train_test_split() : parameter (x , y , test_size= , random_state)
x_train, x_test , y_train, y_test = train_test_split(x,y , test_size=0.4 ,random_state=4)
```

```python
x_train.head()
x_train.shape
```

```python
x_test.head()
```

```python
## step 1: Load regression model from scikit learn
from sklearn.linear_model import LinearRegression
## step 2 : create a variable(model) of Linear Regression type
model = LinearRegression()
# step 3 : fit training data vriables ( x_train , y_train ) in model
model.fit(x_train,y_train)
# step 4 : load predict function ( it will accept test variable)
pred =model.predict(x_test)
# predicted values will be compare with Actual value (y_test)
```

```python
# i will store the independent variable + Actual + predicted value
# x_test + y_test + pred
result = x_test
```

```python
result['Actual'] = y_test
result['Predicted']= pred
```

```python
result
```

```python
result.to_csv('boston_predicted.csv')
```

```python
### Compare Actual Vs Predicted in Regression Graph
plt.figure(figsize=(14,12))
for i in range(len(col)-1):
 plt.subplot(4,4,i+1)
 sns.regplot(x= col[i], y='Actual',data=result , label='Actual')
 sns.regplot(x= col[i], y='Predicted',data=result , label='Predicted')
 plt.legend() # this will create a box where the information which you put in label will shown
plt.show()
```
